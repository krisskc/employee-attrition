# feature engineering

```python
features=Emp_data.loc[:,['EnvironmentSatisfaction','JobInvolvement','JobSatisfaction','OverTime','StockOptionLevel','TotalWorkingYears','YearsInCurrentRole','YearsWithCurrManager']]
labels=Emp_data.iloc[:,1]
```

```python
features
```

```python
labels
```

# data normalization

```python
from sklearn import preprocessing
```

```python
minmax=preprocessing.MinMaxScaler(feature_range=(0,1))
minmax.fit(features)
```

```python
minmax.transform(features)
```

```python
X=features
y=labels
```

```python
from sklearn import neighbors, model_selection
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
import statsmodels.api as sm
model1=sm.OLS(y,X).fit()
model1.summary()
```

```python
X_train_std=minmax.fit_transform(X_train)
X_test_std=minmax.fit_transform(X_test)
```

# logistic regression with test size=0.2

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score,cross_val_predict
lr1=LogisticRegression()
lr1.fit(X_train,y_train)
lr1_acc=cross_val_score(lr1,X_train_std,y_train,cv=3,scoring='accuracy',n_jobs=-1)
```

```python
lr1_proba=cross_val_predict(lr1,X_train_std,y_train,cv=3,method='predict_proba')
lr1_scores=lr1_proba[:,1]
```

```python
lr1_acc
```

```python
y_predict=lr1.predict(X_test)
```

```python
#accuracy score using sklearn
from sklearn.metrics import accuracy_score,recall_score,roc_auc_score,confusion_matrix
```

```python
print('accuracy : %f'%(accuracy_score(y_test,y_predict)*100))
```

```python
print('roc score : %f'%(roc_auc_score(y_test,y_predict)*100))
```

```python
print('recall score : %f'%(recall_score(y_test,y_predict)*100))
```

```python
print(confusion_matrix(y_test,y_predict))
```

```python
#testing probability
probas=lr1.predict_proba(X_test)
```

```python
plt.figure(dpi=150)
plt.hist(probas,bins=20)
plt.title('classification probabilities')
plt.xlabel('probability')
plt.ylabel('# of instances')
plt.xlim([0.5,1.0])
plt.legend(y_test)
plt.show()
```

```python
from sklearn.metrics import roc_curve,roc_auc_score
def roc_cur(title,y_train,scores,label=None):
    fpr,tpr,thresholds=roc_curve(y_train,scores)
    print('AUC curve ({}):{:.2f}'.format(title,roc_auc_score(y_train,scores)))
    plt.figure(figsize=(8,6))
    plt.plot(fpr,tpr,linewidth=2,label=label)
    plt.xlabel('false positive rate',fontsize=16)
    plt.ylabel('true positive rate',fontsize=16)
    plt.title('roc curve:{}'.format(title),fontsize=16)
    plt.show()
```

```python
roc_cur("Logistic Regression",y_train,lr1_scores)
```
